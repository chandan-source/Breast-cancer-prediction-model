cleaning of the dataset

```python
li=['id','Unnamed: 32']
df.drop(li,axis=1,inplace=True)
#removing 2 column from the datset
```

```python
x=df.iloc[:,1:31].values
y=df.iloc[:,0:1].values
#converting dataframes into array
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#splitting the dataset for taining and testing set
```

```python
x_train.size,x_test.size
```

scaling of the dataset


```python
from sklearn.preprocessing import StandardScaler
ssx=StandardScaler()
ssx.fit_transform(x_train)
ssx=StandardScaler()
ssx.fit_transform(x_test)
```

labeling of the string column int a int so for better understandable of system


```python
from sklearn.preprocessing import LabelEncoder
lex=LabelEncoder()
y_train=lex.fit_transform(y_train)
y_test=lex.fit_transform(y_test)
```

```python
y_train,y_test
```

using method svr to tain and test dataset

```python
from sklearn.svm import SVR
SX=SVR(gamma=1)
SX.fit(x_train,y_train)
```

```python
print("taining accuracy:",SX.score(x_train,y_train)*100)
```

```python
SX.fit(x_test,y_test)
print("testing accuracy :",SX.score(x_test,y_test)*100)
```
